```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(0,101,50)
y=np.random.randint(0,101,50)
```

```python
plt.scatter(x,y,marker='+',color="blue")
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([x, y], tick_labels=['x', 'y'])
plt.title('Box Plot')
plt.ylabel('values')
plt.grid(True)
plt.show()
```
